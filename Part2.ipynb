```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.decomposition import PCA
```

```python
data_frame = pd.read_csv("pd_speech_features.csv")
data_frame
```

```python
new_data_frame = data_frame.iloc[:,list(range(1,754))]
new_data_frame
```

```python
test_row =sorted(random.sample(range(1, 756), int(0.3*756)))
```

```python
train_row = []
for i in range(1,len(data_frame)):
    if i not in test_row:
        train_row.append(i)
train_row = sorted(train_row)
```

```python
x_train = data_frame.iloc[train_row, list(range(1,754))]
x_train
```

```python
y_train = data_frame.iloc[train_row, 754]
y_train = y_train.astype(int)
y_train
```

```python
x_test = data_frame.iloc[test_row ,list(range(1,754))]
x_test
```

```python
y_test =  data_frame.iloc[test_row ,754]
y_test =  y_test.astype(int)
y_test
```

```python
p_test = [1 for i in range(226)]
p_train = [1 for i in range(530)]
accuracy_test = sum(y_test.to_numpy() == p_test)/226*100
accuracy_train = sum(y_train.to_numpy() == p_train)/530*100
print("accuracy_test:" , accuracy_test, "\n"+"accuracy_train: " , accuracy_train)
```

# DecisionTree

```python
model = tree.DecisionTreeClassifier(criterion= "entropy", max_depth = 10 ,min_samples_split = 2,min_samples_leaf = 2 , max_features = "log2")
train_model = model.fit(x_train, list(y_train.to_numpy()))
p_train = train_model.predict(list(x_train.to_numpy()))
p_test = train_model.predict(list(x_test.to_numpy()))
accuracy_test = sum(y_test.to_numpy() == p_test)/226*100
accuracy_train = sum(y_train.to_numpy() == p_train)/530*100
print("accuracy_test:" , accuracy_test, "\n"+"accuracy_train: " , accuracy_train)
```

```python
s = cross_val_score(model, x_train, y_train , cv = 6)
c = np.mean(s)
print("cross val score: %" , c*100)
```

```python
m = confusion_matrix(y_test, p_test)
print(m)
```
